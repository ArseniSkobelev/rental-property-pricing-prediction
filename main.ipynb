## Import packages and dataset

```python
import pandas
from sklearn.linear_model import LinearRegression

df = pandas.read_csv('./assets/data/rentals.csv')
```

## Define X and y

```python
X = df[['Area in sqm', 'Number of bedrooms']]
y = df['Monthly price']
```

## Split dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)

# print(X_train.loc[:, "Area in sqm"])
```

## Train model

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

## Test model and visualize result

```python

from matplotlib import pyplot as plt
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

plt.scatter(X_train.loc[:, "Area in sqm"], y_train, color='red', marker='X')
plt.plot(X_train.loc[:, "Area in sqm"], lr.predict(X_train), color='blue')
plt.title('Training set')
plt.xlabel('Area in m2')
plt.ylabel('Monthly price')
plt.show()

plt.scatter(X_test.loc[:, "Area in sqm"], y_test, color='red', marker='X')
plt.plot(X_test.loc[:, "Area in sqm"], lr.predict(X_test), color='blue')
plt.title('Test set')
plt.xlabel('Area in m2')
plt.ylabel('Monthly price')
plt.show()

y_pred = lr.predict(X_test)
print("R^2 : ", r2_score(y_test, y_pred))
print("MAE :", mean_absolute_error(y_test,y_pred))
print("RMSE:",np.sqrt(mean_squared_error(y_test, y_pred)))
```
